```python
#Basado en el demo https://towardsdatascience.com/customer-segmentation-using-k-means-clustering-d33964f238c3

import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

df = pd.read_csv("colores.csv")
df
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df.R, df.G, df.B,c=['#9933FF','#79ECDD','#D1EC79','#F0A44C','#F0624C','#4C5DF0','#32EF5E'],s=50,alpha=1,edgecolor='black')
plt.xlabel("R")
plt.ylabel("G")
ax.set_zlabel('B')
ax.view_init(30, 185)
plt.show()
```

```python
centroides=np.array([[255,0,0],[0,255,0],[0,0,255]], np.int32)
kmeans = KMeans(n_clusters=3,init=centroides,n_init=1)
clusters=kmeans.fit_predict(df)
kmeans.labels_
```

```python
df["label"] = clusters
df
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df.R[df.label==0], df.G[df.label==0], df.B[df.label==0],c='red',s=50,alpha=1,edgecolor='black')
ax.scatter(df.R[df.label==1], df.G[df.label==1], df.B[df.label==1],c='green',s=50,alpha=1,edgecolor='black')
ax.scatter(df.R[df.label==2], df.G[df.label==2], df.B[df.label==2],c='blue',s=50,alpha=1,edgecolor='black')
plt.xlabel("R")
plt.ylabel("G")
ax.set_zlabel('B')
ax.view_init(30, 185)
plt.show()
```
